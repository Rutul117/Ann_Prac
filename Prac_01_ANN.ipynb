```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate x values
x = np.linspace(-10, 10, 100)

sigmoid_y = sigmoid(x)

plt.figure(figsize=(5, 5))
plt.plot(x, sigmoid_y, label='Sigmoid', color='blue')
plt.title("Sigmoid")
plt.show
```

```python
def relu(x):
    return np.maximum(0, x)
relu_y = relu(x)
plt.plot(x, relu_y, label='ReLU', color='red')
plt.title("Relu")
plt.show
```

```python
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
    # return np.tanh(x)

tanh_y = tanh(x)
plt.plot(x, tanh_y, label='Tanh', color='green')
plt.title("Tanh")
plt.show
```

```python
def leaky_relu(x, alpha=0.01):
    return np.maximum(alpha * x, x)
    
leaky_relu_y = leaky_relu(x)
plt.plot(x, leaky_relu_y, label='Leaky ReLU', color='purple')
plt.title("Leaky Relu")
plt.show
```

```python
def step_function(x):
    return np.where(x > 0, 1, 0)

step_function_y = step_function(x)
plt.plot(x, step_function_y, label='Step Function', color='blue')
plt.title("Step Function")
plt.show
```

```python
def softmax(x):
    exp_values = np.exp(x - np.max(x))
    return exp_values / np.sum(exp_values)

softmax_y = softmax(x)
plt.plot(x, softmax_y, label='Softmax Function', color='red')
plt.title("Softmax")
plt.show
```
